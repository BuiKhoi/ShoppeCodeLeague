```python
import os
import numpy as np
```

```python
train_folder = './shopee-product-detection-dataset/train/train/'
val_folder = './shopee-product-detection-dataset/train/val/'
```

```python
for folder_name in os.listdir(train_folder):
    if '.DS' not in folder_name:
        os.mkdir(val_folder + folder_name)
```

```python
move_ratio = 0.3
```

```python
for class_folder in os.listdir(train_folder):
    if '.DS' not in class_folder:
        class_path = train_folder + class_folder + '/'
        destination_path = val_folder + class_folder + '/'
        for image_file in os.listdir(class_path):
            if '.DS' not in image_file:
                rand = np.random.random()
                if rand < move_ratio:
                    os.rename(class_path + image_file, destination_path + image_file)
                    print('Moved {}'.format(class_path + image_file))
```
